## Q2: What's the most common tree species in San Francisco?

```python
# group by species

# group by tree species
sf_trees_species = sf_trees.groupby(['qspecies']).count()
sf_trees_species = sf_trees_species[['treeid']].copy()
sf_trees_species
```

That's a lot! But there are at least three categories that I want to replace with "unknown". I will replace them in the `sf_trees` data frame.
- ::
- :: To Be Determine
- :: Tree
- Tree(s) ::

```python
# replace 4 tree species

sf_trees = sf_trees.replace({
    'qspecies' : 
    {'::' : 'unknown',
     ':: To Be Determine' : 'unknown', 
     ':: Tree' : 'unknown',
     'Tree(s) ::' : 'unknown'}
})
```

```python
# after replacing the species, try groupby again
# group by tree species
sf_trees_species = sf_trees.groupby(['qspecies']).count()
sf_trees_species = sf_trees_species[['treeid']].copy()
sf_trees_species
```

```python
# reset index

sf_trees_species = sf_trees_species.reset_index()
sf_trees_species
```

```python
# rename the columns
sf_trees_species.columns = ['species','treecount']
sf_trees_species
```

```python
# sort by treecount

sf_trees_species = sf_trees_species.sort_values(by='treecount',ascending=False).copy()
sf_trees_species = sf_trees_species.reset_index(drop=True)
sf_trees_species
```

# Download basemap

Download SF neighborhooods shapefile from [DataSF portal]('https://data.sfgov.org/Geographic-Locations-and-Boundaries/SF-Find-Neighborhoods/pty2-tcw4').   
[Download shapefile here.]('https://data.sfgov.org/api/geospatial/pty2-tcw4?method=export&format=Shapefile')   
Rename it `sf_neighborhoods`. 

```python
sf_trees.info()
```

```python
# reformat the data set to only needed columns

sf_trees_map = sf_trees[['treeid','qaddress','latitude','longitude','location']].copy()
sf_trees_map
```

```python
# read shapefile

sf_map = gpd.read_file('sf_neighborhood.zip')
sf_map.to_crs(epsg=4326).plot()
```

```python
# create geopandas dataframe

geo_sf_trees = gpd.GeoDataFrame(
    sf_trees_map, geometry=gpd.points_from_xy(sf_trees_map.longitude, sf_trees_map.latitude)
)
```

```python
geo_sf_trees.head()
```

```python
geo_sf_trees.info()
```

```python


# plot geodataframe on the basemap

geo_sf_trees.plot(marker='*', color='green', markersize=5)
```
